**Linear Algebra**:  


1 - Vector Operations:

$$\textbf{v}_1 = \begin{bmatrix} 1\\2\\3 \end{bmatrix}$$

$$\textbf{v}_2 = \begin{bmatrix} 0\\0\\2 \end{bmatrix}$$

$$\textbf{v}_1 \textbf{v}_2 = \text{ } ?$$

$$\textbf{v}_1^T \textbf{v}_2 = 6$$

$$\textbf{v}_1^T \textbf{v}_2 = \textbf{v}_2^T \textbf{v}_1 \text{ => dot product commutativity}$$ 




2 - Matrix Operations:

$$\textbf{M}_1 = \begin{bmatrix} 1&2\\ 3&4 \end{bmatrix}$$

$$\textbf{M}_2 = \begin{bmatrix} 0&2\\ 0&7 \end{bmatrix}$$

$$\textbf{M}_1 + \textbf{M}_2 = \begin{bmatrix} 1&4\\ 3&11 \end{bmatrix}$$

$$\textbf{M}_1 \textbf{M}_2 = \begin{bmatrix} 0&16\\ 0&34 \end{bmatrix}$$

$$\textbf{M}_1 \odot \textbf{M}_2 = \begin{bmatrix} 0&4\\ 0&28 \end{bmatrix}$$

3 - Matrix Product Properies: Let $\textbf{M}_1, \textbf{M}_2, \text{and } \textbf{M}_3$ be matrices.

$$\textbf{M}_1 \left( \textbf{M}_2 + \textbf{M}_3 \right) = \textbf{M}_1 \textbf{M}_2 + \textbf{M}_1 \textbf{M}_3 \text{ => distributivity}$$

$$\textbf{M}_1 \left( \textbf{M}_2 \textbf{M}_3 \right) = \left( \textbf{M}_1 \textbf{M}_2 \right) \textbf{M}_3 \text{ => associativity}$$

$$\textbf{M}_1 \textbf{M}_2 \neq \textbf{M}_2 \textbf{M}_1 \text{, in general}$$ 

$$\textbf{M}_1 \textbf{M}_1^{-1} = \textbf{I}$$

$$\left( \textbf{M}_1 \textbf{M}_2 \right)^{-1} = \textbf{M}_2^{-1} \textbf{M}_1^{-1}$$

$$\left( \textbf{M}_1 \textbf{M}_2 \right)^{T} = \textbf{M}_2^{T} \textbf{M}_1^{T}$$

4 - Linear Transformation of a vector:

$$\textbf{A} \textbf{x} = \textbf{b}, A \in R^{n x n}, b \in R^n$$

*Can solve for x using inversion or Gaussian elimination*:
$$\textbf{x} = \textbf{A}^{-1}\textbf{b} $$

5 - Norms: Measuring the *size* of a vector or matrix. Let $\textbf{x},\textbf{z} \in R^n$, $\textbf{A} \in R^{mxn}$.

$$||\textbf{x}||_p = \left( \sum_{i=1}^n |x_i|^p \right)^\frac{1}{p}$$

$$||\textbf{x}||_2 = \left( \sum_{i=1}^n x_i^2 \right)^\frac{1}{2}$$

$$||\textbf{x}||_1 = \sum_{i=1}^n |x_i|$$

$$||\textbf{x}||_{\infty} = \max_{{i}} |x_i|$$

$$\text{dist}(\textbf{x},\textbf{z}) = ||\textbf{x} - \textbf{z}||_2$$

$$||\boldsymbol{A}||_{F} = \left( \sum_{i=1}^m \sum_{j=1}^n A_{i,j}^2 \right)^\frac{1}{2} $$

$$||\boldsymbol{A}||_{F} = \left( \text{Tr}(\boldsymbol{A}^T \boldsymbol{A}) \right)^\frac{1}{2}, \text{ where Tr}(\boldsymbol{M}) = \sum_{i=1}^n M_{i,i} $$



6 - Angle between vectors. Let $\textbf{x},\textbf{z} \in R^n$.

$$\textbf{x}^T \textbf{z} = ||\textbf{x}||_2 ||\textbf{z}||_2 \text{ cos}(\theta)$$

$$\text{similarity} = \text{cos}(\theta) = \frac{\textbf{x} \cdot \textbf{z}}{||\textbf{x}||_2 ||\textbf{z}||_2}$$


7 - Symmetric matrix. Let $\textbf{A} \in R^{nxn}$.

$$\textbf{A} = \textbf{A}^T$$


8 - Special vectors and matrices. Let $\textbf{x},\textbf{z} \in R^n$.

$$\text{Unit vector: } ||\textbf{x}||_2 = 1$$

$$\text{Orthogonal vector: } \textbf{x}^T \textbf{y} = 0$$

$$\text{Orthonormal vectors: vectors are orthogonal and have unit norm}$$

$$\text{Orthogonal matrix: } \textbf{A}^{-1} = \textbf{A}^T$$



9 - Matrix Decomposition (1). Let $\textbf{v} \in R^n$, $\textbf{A} \in R^{nxn}$. 

$$\textbf{A} = \textbf{V} \boldsymbol{\Lambda} \textbf{V}^{-1} \text{ where } \boldsymbol{\Lambda} = \text{diag}(\boldsymbol{\lambda}), \boldsymbol{\Lambda} \text{ is an eigenvector matrix, and }\boldsymbol{\lambda}\text{ is an eigenvalue vector.}$$ 

$$\boldsymbol{Av}=\boldsymbol{\lambda v}$$

$$\boldsymbol{Av}-\boldsymbol{\lambda v} = \left( \boldsymbol{A}-\boldsymbol{\lambda I_n} \right) v = 0$$

$$\text{det}(\boldsymbol{A}-\boldsymbol{\lambda I_n}) = |\boldsymbol{A}-\boldsymbol{\lambda I_n}| = 0 \text{ for a non-zero solution.}$$

Eigenvalues and eigenvectors will be *real* if $\boldsymbol{A}$ is *real* and *symmetric*;  $\textbf{V}$ will be an orthogonal matrix.

If any eigenvalue is $0$, the matrix $\boldsymbol{A}$ is singular, or non-invertible. 

A Positive-Definite (PD) matrix has all positive eigenvalues. 

A Positive Semi Definite (PSD) matrix has non-negative eigenvalues. 


10 - Matrix Decomposition (2). Let $\textbf{A} \in R^{mxn}$. 

$$\boldsymbol{A} = \boldsymbol{U} \boldsymbol{D} \boldsymbol{V}^T \text{ }\left[\text{SVD} - \text{ more general than Eigendecomposition}\right]$$

$\boldsymbol{U}$ and $\boldsymbol{V}$ - orthogonal matrices, $\boldsymbol{D}$ - diagonal matrix (not necessarily square).

Elements on diagonal of $\boldsymbol{D}$ - singular values of $\boldsymbol{A}$

Columns of $\boldsymbol{U}$ - left singular vectors

Columns of $\boldsymbol{V}$ - right singular vectors

Left singular vectors of $\boldsymbol{A}$ - eigenvectors of $\boldsymbol{A} \boldsymbol{A}^T$

Right singular vectors of $\boldsymbol{A}$ - eigenvectors of $\boldsymbol{A}^T \boldsymbol{A}$

Nonzero singular values of $\boldsymbol{A}$ - squareroot of eigenvalues of $\boldsymbol{A}^T \boldsymbol{A}$ or, equivalently, $\boldsymbol{A} \boldsymbol{A}^T$. 


## ML pipeline

The machine learning pipeline refers to the end-to-end process of getting raw data, building a model, deploying it, and updating it over time. It typically consists of 5 main stages as shown in the figure below (source: https://www.oreilly.com/library/view/machine-learning-with/9781783288519/graphics/8519OS_02_02.jpg).

[scikit-learn](http://scikit-learn.org) and [pandas](https://pandas.pydata.org/) are two very popular Python libraries used for machine learning solutions. We will be writing code to create a simple training pipeline using these two libraries.

```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

**Step 1** Data Acquisition: This is the stage where raw data is collected.

pandas contains several functions to read the raw data available in formats such as csv, excel, json, txt, etc. The data read in is stored as a DataFrame, which is like a 2-dimensional tabular object from which data can be accessed by row and column names or indices

```python
# read in the toy dataset
df = pd.read_csv('boston_housing.csv')

# this will give a sneak peek of the first 7 rows of the DataFrame
df.head(n=7)
```

```python
# HW2 HINT - values in column can be accessed like this
age_values = df['AGE'].values

# split the dataset into X and y matrices
X = df.drop('MEDV', axis=1)
y = df[['MEDV']]

# this further splits the X and y into two subsets:
# one used for training model and other used to test how well the model has been learned 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

**Step 2** Preprocessing and Feature Engineering: In this stage, raw data is transformed into a form that is suitable for training models. This includes things like removing NaN valued rows, removing less useful features, scaling and centering, etc. 

```python
# center and scale raw data values
X_train_cs = (X_train - X_train.mean()) / X_train.std()
X_train_cs.head()
```

**Step 3** Train the model!

```python
# HW 2 HINT - fit and ordinary least squares model
ols_model = LinearRegression()
ols_model.fit(X_train_cs, y_train)

# HW2 HINT - after the model has been train, you can find out the coefficients (i.e. the "W" vector) ike this
coeffs = ols_model.coef_
```

**Step 4** Testing the model.

You now have a trained model trained. But how good is it, really? We find that out by using our model to run inference on the test dataset created earlier. Based on the results we get, we may then decide to fine tune training hyperparameters (e.g. lambda in case of Ridge regression) to see if we can build a better model using a different value of the hyperparameter.

There are several metrics used to evaluate the performance of a model. One such popular metric is mean squared error

```python
# dont forget to apply same transformation to test data as was applied to the train data
X_test_cs = (X_test - X_test.mean()) / X_test.std()

# calculate mean squared error
mse = mean_squared_error(y_pred=ols_model.predict(X_test_cs),
                        y_true=y_test)
print(mse)
```

**Step 5** Deploying the model

This is the stage where we release the model into the real world. In most cases, the model is retrained as more and more new data is collected.
